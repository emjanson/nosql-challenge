5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Define a function to update latitude and longitude fields
def update_geocode_to_decimal(collection):
    try:
        collection.update_many(
            {},
            [
                {'$set': {'geocode.latitude': {'$toDouble': '$geocode.latitude'}}},
                {'$set': {'geocode.longitude': {'$toDouble': '$geocode.longitude'}}}
            ]
        )
        print("Latitude and longitude data type update successful.")
    except Exception as e:
        print(f"Latitude and longitude data type update failed: {str(e)}")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Define a function to update RatingValue to integers
def update_rating_to_integer(collection):
    try:
        collection.update_many(
            {'RatingValue': {'$exists': True, '$ne': None}},
            [{'$set': {'RatingValue': {'$toInt': '$RatingValue'}}}]
        )
        print("RatingValue data type update successful.")
    except Exception as e:
        print(f"RatingValue data type update failed: {str(e)}")
```

```python
# Update latitude and longitude fields for each document
update_geocode_to_decimal(establishments)
update_rating_to_integer(establishments)
```

```python
# Check that the coordinates and rating value are now numbers
def verify_data_types(collection):
    for document in collection.find():
        # Check 'geocode' field
        try:
            latitude = float(document['geocode']['latitude'])
            longitude = float(document['geocode']['longitude'])
        except (KeyError, ValueError):
            print(f"Invalid 'geocode' data type in document {_id}: {document}")

        # Check 'RatingValue' field
        if 'RatingValue' in document:
            rating_value = document['RatingValue']
            if rating_value is not None:
                try:
                    int(rating_value)
                except ValueError:
                    print(f"Invalid 'RatingValue' data type in document {_id}: {document}")

    print("Data type verification complete. Check for any error messages printed.")

verify_data_types(establishments)
```
